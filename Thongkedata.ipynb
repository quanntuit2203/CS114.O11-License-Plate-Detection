**3.** Đếm số bounding box tối đa được gán trong tập train

```python
def find_max_boundingbox(folder_path):
    bb_count = 0

    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)

            with open(file_path, 'r') as file:
                lines = file.readlines()

            line_count = len(lines)
            bb_count = max(bb_count, line_count)

    print(f"Số bounding box tối đa trong tập train có được là: {bb_count}")

# Sử dụng hàm với đường dẫn của thư mục chứa các file txt
folder_path = '/content/drive/MyDrive/Tester/train/labels'
find_max_boundingbox(folder_path)
```

**4.** Từ 1 đến 8 bounding box được tạo, đếm số lượng ảnh có được ứng với số lượng bounding box

```python
def count_pic_in_eachbb(folder_path):
    # Khởi tạo các biến đếm
    One_BB = 0
    Two_BB = 0
    Three_BB = 0
    Four_BB = 0
    Five_BB = 0
    Six_BB = 0
    Seven_BB = 0
    Eight_BB = 0

    # Lặp qua tất cả các file trong thư mục
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)

            # Mở file và đọc nội dung
            with open(file_path, 'r') as file:
                lines = file.readlines()

            # Đếm số lượng dòng và tăng giá trị biến tương ứng
            for line_num, line in enumerate(lines, start=1):
                if line_num <= 9:
                    if line.strip():
                        if line_num == 1:
                            One_BB += 1
                        elif line_num == 2:
                            Two_BB += 1
                        elif line_num == 3:
                            Three_BB += 1
                        elif line_num == 4:
                            Four_BB += 1
                        elif line_num == 5:
                            Five_BB += 1
                        elif line_num == 6:
                            Six_BB += 1
                        elif line_num == 7:
                            Seven_BB += 1
                        elif line_num == 8:
                            Eight_BB += 1


    # In kết quả
    print('One_bounding box:', One_BB)
    print('Two_bounding boxes:', Two_BB)
    print('Three_bounding boxes:', Three_BB)
    print('Four_bounding boxes:', Four_BB)
    print('Five_bounding boxes:', Five_BB)
    print('Six_bounding boxes:', Six_BB)
    print('Seven_bounding boxes:', Seven_BB)
    print('Eight_bounding boxes:', Eight_BB)

# Sử dụng hàm với đường dẫn của thư mục chứa các file txt
folder_path = '/content/drive/MyDrive/Tester/train/labels'
count_pic_in_eachbb(folder_path)
```

**5.** Đưa ra 1 file txt mẫu có số bounding box lớn nhất và chi tiết thông số của từng bounding box có trong file đó

```python
def find_file_with_max_boundingbox(folder_path):
    max_line_count = 0
    max_line_content = ""
    max_line_file = ""

    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)

            with open(file_path, 'r') as file:
                lines = file.readlines()

            line_count = len(lines)
            if line_count > max_line_count:
                max_line_count = line_count
                max_line_content = "".join(lines)
                max_line_file = filename

    print(f"Max Bounding Box Count: {max_line_count}")
    print(f"File with Max Bounding Box: {max_line_file}")
    print(f"Max Bounding Box Content:\n{max_line_content}")

    if max_line_count > 0:
        with open('max_bbox_content.txt', 'w') as output_file:
            output_file.write(max_line_content)

# Sử dụng hàm với đường dẫn của thư mục chứa các file txt
folder_path = '/content/drive/MyDrive/Tester/train/labels'
find_file_with_max_boundingbox(folder_path)
```
